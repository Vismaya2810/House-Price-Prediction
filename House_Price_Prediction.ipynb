```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/USA_Housing.csv') # Dataset: https://www.kaggle.com/datasets/aariyan101/usa-housingcsv
data
```

```python
data.describe()
```

```python
X = pd.DataFrame(data,columns=['House_Age'])
y = pd.DataFrame(data,columns=['House_Price'])
```

```python
regression = LinearRegression()
regression.fit(X,y)
```

```python
# Slope Coefficient
regression.coef_
```

```python
# Intercept
regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y,alpha=0.3)
plt.plot(X,regression.predict(X),color='red',linewidth=4)
plt.title('House Age vs House Price')
plt.xlabel('House Age')
plt.ylabel('House Price')
plt.show()
```

```python
regression.score(X,y)
```
